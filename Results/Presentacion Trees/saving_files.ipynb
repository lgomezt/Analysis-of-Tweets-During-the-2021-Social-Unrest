```python
# Mathematical and Data Managment
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

# Data Visualization
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter
from IPython.display import HTML

# Settings for plots
plt.rcParams['font.family'] = 'cmr10'
#plt.rcParams['mathtext.fontset'] = 'cm'

#sns.set_style('white')
```

```python
group_segregation_daily = pd.read_pickle('/mnt/disk2/Data/Pickle/Segregation/group_segregation_daily.pkl')
global_segregation_daily = pd.read_pickle('/mnt/disk2/Data/Pickle/Segregation/global_segregation_daily.pkl')

group_segregation_3day = pd.read_pickle('/mnt/disk2/Data/Pickle/Segregation/group_segregation_3day.pkl')
global_segregation_3day = pd.read_pickle('/mnt/disk2/Data/Pickle/Segregation/global_segregation_3day.pkl')

individual_group_segregation = pd.read_pickle('/mnt/disk2/Data/Pickle/Segregation/individual_group_segregation.pkl')
individual_node_segregation = pd.read_pickle('/mnt/disk2/Data/Pickle/Segregation/individual_node_segregation.pkl')

# Resetting index to make 'Date' a column for Seaborn's use
df_daily = group_segregation_daily.reset_index()
sin_inclasificados_daily = df_daily[df_daily['Political Label'] != 'Sin Clasificar']

df_3day = group_segregation_3day.reset_index()
sin_inclasificados_3day = df_3day[df_3day['Political Label'] != 'Sin Clasificar']
```

```python
fig, ax = plt.subplots(figsize=(12, 5))  # Increased size from (12,5) to (16,8)
# Enable the use of Font
plt.rcParams['axes.formatter.use_mathtext'] = True
# Plot
sns.lineplot(data=df_daily, x='Date', y='Proximity to Others', hue='Political Label',
             palette=["forestgreen", "red", "darkblue", "purple"], ax=ax, marker='o')

ax.xaxis.set_major_locator(mdates.DayLocator(interval=6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
ax.tick_params(axis = 'x', rotation = 45)
ax.set_ylabel('Proximity to Others Index')
ax.set_xlabel('Daily basis')
plt.legend(title='Political Label', bbox_to_anchor=(1, 1), loc='best')
fig.savefig("Proximity_to_others_con_inclasificados_daily.png", dpi=300, bbox_inches='tight')
plt.show()
```
